```python
import pickle
import matplotlib.pyplot as plt
```

```python
history_initial = pickle.load(open('history_djia30.pkl', 'rb'))
history_initial['train_loss'] = history_initial['train_loss'][:30]
history_initial['accuracy'] = history_initial['accuracy'][:30]
history_initial['precision'] = history_initial['precision'][:30]
history_initial['recall'] = history_initial['recall'][:30]
history_initial['f1'] = history_initial['f1'][:30]
history_initial['auc'] = history_initial['auc'][:30]

history1 = pickle.load(open('fold_1_history.pkl', 'rb'))
history2 = pickle.load(open('fold_2_history.pkl', 'rb'))
history3 = pickle.load(open('fold_3_history.pkl', 'rb'))
history4 = pickle.load(open('fold_4_history.pkl', 'rb'))
history5 = pickle.load(open('fold_5_history.pkl', 'rb'))


fold_histories = [history1, history2, history3, history4, history5]

metrics = {
    'Loss': 'train_loss',
    'Accuracy': 'accuracy',
    'Precision': 'precision',
    'Recall': 'recall',
    'F1 Score': 'f1',
    'AUC': 'auc'
}

for title, key in metrics.items():
    plt.figure(figsize=(6,4))
    for idx, hist in enumerate(fold_histories, start=1):
        plt.plot(hist[key], marker='o', label=f'Fold {idx}')
    plt.plot(history_initial[key], linestyle='--', marker='x', label='Direct')
    plt.title(f'{title} vs Epoch', fontsize=14)
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel(title, fontsize=12)
    plt.legend(fontsize=10, loc='best')
    plt.tight_layout()
    plt.show()
```
